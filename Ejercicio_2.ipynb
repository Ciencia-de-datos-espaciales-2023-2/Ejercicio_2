```python
Laureates=pd.read_html(LaureatesLink,header=0,flavor='bs4')
Laureates_DF=Laureates[0]

#Eliminando las columnas "Image" y "Ref"
Laureates_DF.drop(columns=["Image","Ref"], inplace=True)

# Renombrando las columnas
Laureates_DF.rename(columns={"Laureate[A]":"Laureate","Country[B]":"Country","Rationale[C]":"Rationale"},inplace=True)

#Creando dos nuevas columnas (Birth_year y Death_year a partir de datos preexistentes de la columna Laureate)
Laureates_DF_2=Laureates_DF["Laureate"].str.split("(", expand=True) 
Laureates_DF["Laureate"]=Laureates_DF_2[0]
##Eliminando "b." y ")"
Laureates_DF_2[1]=Laureates_DF_2[1].str.replace(pat=r'(\Ab.)',repl='',regex=True)
Laureates_DF_2[1]=Laureates_DF_2[1].str.replace(pat=')',repl='')
Laureates_DF_2=Laureates_DF_2[1].str.split("–", expand=True)

#Columnas resultantes luego de limpiar
Laureates_DF_2
```

```python
#Reemplazando valores "None" en la segunda columna con "Alive"
Laureates_DF_2[1]=Laureates_DF_2[1].replace({None:'Alive'})
```

```python
#Dataframe resultante
Laureates_DF["Birth_year"]=Laureates_DF_2[0]
Laureates_DF["Death_year"]=Laureates_DF_2[1]
Laureates_DF
```
